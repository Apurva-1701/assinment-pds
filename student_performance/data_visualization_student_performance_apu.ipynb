```python
import matplotlib.pyplot as plt
import seaborn as sns
output_dir = 'visualizations_alt/'
os.makedirs(output_dir, exist_ok=True)
sns.set(style="darkgrid")
def save_and_show_plot(fig, filename):
    fig.savefig(os.path.join(output_dir, filename))
    plt.show()

```

```python

plt.figure(figsize=(10,6))
sns.histplot(data[['math score', 'reading score', 'writing score']], bins=20, kde=True, element="step", stat="density")
plt.title('Distribution of Math, Reading, and Writing Scores')
plt.xlabel('Score')
plt.ylabel('Density')
save_and_show_plot(plt.gcf(), 'distribution_scores.png')

```

```python



plt.figure(figsize=(12,8))
sns.boxplot(x='gender', y='math score', data=data, palette="Set3", width=0.5)
sns.boxplot(x='gender', y='reading score', data=data, palette="Set3", width=0.5)
sns.boxplot(x='gender', y='writing score', data=data, palette="Set3", width=0.5)
plt.title('Box Plot of Scores by Gender')
save_and_show_plot(plt.gcf(), 'boxplot_scores_by_gender.png')

```

```python

sns.pairplot(data, vars=['math score', 'reading score', 'writing score'], hue='gender', palette='Set2', markers=["o", "s"])
plt.suptitle('Pair Plot of Math, Reading, and Writing Scores', y=1.02)
plt.savefig(os.path.join(output_dir, 'pairplot_scores_gender.png'))
plt.show()

```

```python

plt.figure(figsize=(8,6))
sns.countplot(x='lunch', data=data, palette='pastel')
plt.title('Count Plot of Students by Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('Count')
save_and_show_plot(plt.gcf(), 'countplot_lunch_type.png')

```

```python

plt.figure(figsize=(12,8))
sns.violinplot(x='gender', y='math score', hue='lunch', data=data, split=True, inner="quart", palette="Set1")
plt.title('Violin Plot of Math Scores by Gender and Lunch Type')
save_and_show_plot(plt.gcf(), 'violinplot_math_gender_lunch.png')

```

```python

plt.figure(figsize=(8,6))
corr = data[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Scores')
save_and_show_plot(plt.gcf(), 'heatmap_correlation_scores.png')
```
